```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Ventas_totales.csv")
df
```

Reemplazar valores nulos en cada columna

```python
#Identificar valores nulos por columna
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos por dataframe
valores_nulos=df.isnull().sum().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
df2=df.copy()
```

```python
df2.head(5)
```

```python
#En mi analisis exploratorio me di cuenta que el salon de ventas era resultado de 
#ventas_totales_canal_venta m√°s los canales_on_line entonces eso fue lo que realice 
df2["salon_ventas"] = df2["salon_ventas"].fillna(df2["ventas_totales_canal_venta"] + df2["canales_on_line"])
df2
```

```python
#Corroboramos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```

**En este caso nos dimos cuenta que la resta de los valores era el resultado de los valores nulos que hac√≠an falta dependiendo el metodo de pago**

```python
df2["tarjetas_debito"] = df2["tarjetas_debito"].fillna(df2["ventas_totales_medio_pago"] - df2["tarjetas_credito"] - df2["otros_medios"] - df2["efectivo"])
df2
```

```python
df2["tarjetas_credito"] = df2["tarjetas_credito"].fillna(df2["ventas_totales_medio_pago"] - df2["tarjetas_debito"] - df2["otros_medios"] - df2["efectivo"])
df2
```

```python
df2["otros_medios"] = df2["otros_medios"].fillna(df2["ventas_totales_medio_pago"] - df2["tarjetas_debito"] - df2["tarjetas_credito"] - df2["efectivo"])
df2
```

```python
#Corroboramos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```

**En este caso el valor que nos pedia era la suma de todos los productos de alimentos y bebidas**

```python
df2["subtotal_ventas_alimentos_bebidas"] = df2["subtotal_ventas_alimentos_bebidas"].fillna(df2["bebidas"] + df2["almacen"] + df2["panaderia"] + df2["lacteos"] + df2["carnes"]+ df2["verduleria_fruteria"])
df2
```

```python
#Corroboramos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```

**En este caso el valor que nos pedia era la resta de subtotal_ventas_alimentos_bebidas y de caulquier alimento o bebida y de esta manera encontrariamos el faltante**

```python
df2["bebidas"] = df2["bebidas"].fillna(df2["subtotal_ventas_alimentos_bebidas"] - df2["almacen"] - df2["panaderia"] - df2["lacteos"] - df2["carnes"]- df2["verduleria_fruteria"])
df2
```

```python
df2["almacen"] = df2["almacen"].fillna(df2["subtotal_ventas_alimentos_bebidas"] - df2["bebidas"] - df2["panaderia"] - df2["lacteos"] - df2["carnes"]- df2["verduleria_fruteria"])
df2
```

```python
df2["panaderia"] = df2["panaderia"].fillna(df2["subtotal_ventas_alimentos_bebidas"] - df2["bebidas"] - df2["almacen"] - df2["lacteos"] - df2["carnes"]- df2["verduleria_fruteria"])
df2
```

```python
df2["lacteos"] = df2["lacteos"].fillna(df2["subtotal_ventas_alimentos_bebidas"] - df2["bebidas"] - df2["almacen"] - df2["panaderia"] - df2["verduleria_fruteria"])
df2
```

```python
df2["carnes"] = df2["carnes"].fillna(df2["subtotal_ventas_alimentos_bebidas"] - df2["bebidas"] - df2["almacen"] - df2["panaderia"] - df2["verduleria_fruteria"]- df2["lacteos"])
df2
```

```python
#Corroboramos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```

**En este caso rellenemos los faltantes con un cero ya que no existia una formula para llegar a ese n√∫mero**

```python
df2["verduleria_fruteria"]= df2["verduleria_fruteria"].fillna(0) 
df2
```

```python
df2["alimentos_preparados_rotiseria"]= df2["alimentos_preparados_rotiseria"].fillna(0) 
df2
```

```python
df2["indumentaria_calzado_textiles_hogar"]= df2["indumentaria_calzado_textiles_hogar"].fillna(0) 
df2
```

```python
df2["electronicos_articulos_hogar"]= df2["electronicos_articulos_hogar"].fillna(0) 
df2
```

```python
df2["otros"]= df2["otros"].fillna(0) 
df2
```

```python
#Corroboramos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```
